This dataset has 6 variables. There appears to be an ID for each observation, measurements for the length and the width of the sepals and petals, and the species of the flower. The sepal is the leaf that accompanies the flower, while the petal is the actual leaf of the flower. Next, I'd want to see the different types of species featured in this dataset.

```python
iris.Species.unique() #find the names of each species featured in the dataset
```

There appear to be three different species of flowers featured in this dataset. I would be interested to see some summary statistics by species and to compare the differences.

```python
iris.groupby("Species").describe() #summary statistics for each species of flower.
```

Now we have some summary statistics, but we don't see the whole story just yet. I think some scatterplots by species would help us visualize the relationships between certain variables by each species.

```python
fig = iris[iris.Species=='Iris-setosa'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='orange', label='Setosa')
iris[iris.Species=='Iris-versicolor'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='blue', label='versicolor',ax=fig)
iris[iris.Species=='Iris-virginica'].plot(kind='scatter',x='SepalLengthCm',y='SepalWidthCm',color='green', label='virginica', ax=fig)
fig.set_xlabel("Sepal Length")
fig.set_ylabel("Sepal Width")
fig.set_title("Sepal Length VS Width")
fig=plt.gcf()
fig.set_size_inches(12, 8)
plt.show()
```

```python
fig = iris[iris.Species=='Iris-setosa'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='orange', label='Setosa')
iris[iris.Species=='Iris-versicolor'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='blue', label='versicolor',ax=fig)
iris[iris.Species=='Iris-virginica'].plot.scatter(x='PetalLengthCm',y='PetalWidthCm',color='green', label='virginica', ax=fig)
fig.set_xlabel("Petal Length")
fig.set_ylabel("Petal Width")
fig.set_title(" Petal Length vs. Width")
fig=plt.gcf()
fig.set_size_inches(12, 8)
plt.show()
```
